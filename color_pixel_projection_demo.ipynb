# Color pixel projection on color axis

This code provides functions and examples on how to project the color pixels of an image onto one dominant color axis, e.g., orange-teal or green-purple.

First, let's import the required functions and libraries that we are going to use in the code. 

```python
import os

import numpy as np
import cv2

from HSV_manipulations import get_HSV_RGB
from color_pixel_projection import recolor_image, plot_color_wheel, plot_color_axis
from utils import imshow
```

We define the path of the image to modify and the path of the directory where we will save the results.

```python
img_dir = "images/" # Image directory
img_file = "nicholas-roerich_tent-mountain-1933.jpg" # Image file to modify
results_dir = "results/image_recoloring/color_pixel_projections/" # Where to save the altered images
os.makedirs(results_dir, exist_ok=True) # Creates results_dir if it does not exist
```

```python
""" Here we read and plot the original image that we wish to modify."""
bgr_img = cv2.imread(img_dir + img_file) # cv2 encodes images in BGR
rgb_img = bgr_img[:, :, ::-1] # Converts BGR to RGB
imshow(rgb_img, img_file)
```

```python
""" Here we plot its pixel distribution on the color wheel."""
hsv_img, rgb_img = get_HSV_RGB(img_dir + img_file, scale_factor=1)

# Downsample pixels for efficiency
downsampling_factor = 10
plot_color_wheel(hsv_img, results_dir+'colorwheel.jpg', downsampling_factor=downsampling_factor)
```

### Color mapping to orange-teal axis

Below we perform image recoloring by moving the pixels onto axis 30-210 degrees (orange-teal).
The parameter `rescale_shift` determines the extent of the transformation. If `rescale_shift=1`, the pixels are mapped exactly on the axis. Larger values move the pixels closer to the axis.
The resulting image and new colorwheel are displayed and saved in the `results_dir`.



```python
# Here, we plot the orange-teal axis
plot_color_axis(30, results_dir, n_pixels=100)
```

```python
# Here, we modify the image
hue_degree = 30
rescale_shift = 1
recolor_image(hsv_img, hue_degree, rescale_shift, img_file, results_dir)
```

### Changing `rescale_shift`

```python
rescale_shift = 1.2
recolor_image(hsv_img, hue_degree, rescale_shift, img_file, results_dir)
```

```python
rescale_shift = 1.5
recolor_image(hsv_img, hue_degree, rescale_shift, img_file, results_dir)
```

### Color mapping to green-magenta axis
Here, we perform the same mapping but onto the green-magenta axis (degrees 120-300 on the color circle).

```python
hue_degree = 120
rescale_shift = 1
recolor_image(hsv_img, hue_degree, rescale_shift, img_file, results_dir)
```

```python
hue_degree = 120
rescale_shift = 1.2
recolor_image(hsv_img, hue_degree, rescale_shift, img_file, results_dir)
```

```python
hue_degree = 120
rescale_shift = 1.5
recolor_image(hsv_img, hue_degree, rescale_shift, img_file, results_dir)
```

### Sanity check
For high values of rescale_shift (e.g., 1000), the image remains the same as the shifts become very small.

```python
# Here, we modify the image
hue_degree = 30
rescale_shift = 1000
recolor_image(hsv_img, hue_degree, rescale_shift, img_file, results_dir)
```
